# Transfer alignments from NIV84 to new NIV11 target tokens

Our NIV11 alignment data has numerous issues: incorrect target tokens, and missing alignments. Rather than try to fix it, I'm starting fresh.

1. Align NIV84 targets to NIV11 targets from most recent kathairo data. Many verses should be identical.
2. Identify simple equivalencies ("desert" -> "wilderness"), build a mapping table, and extend the target alignment to include these comparable terms.
3. Transfer alignments directly wherever the text is identical or equivalent.
4. Identify the residue and make a plan.

The latest kathairo'ed version: https://github.com/Clear-Bible/alignments-eng/data/NIV11/target_NIV11.tsv (and OT/NT versions). 


```python
from importlib import reload

from biblealignlib.burrito import CLEARROOT, AlignmentSet, BaseToken, DiffRecord, Manager, Target, target
# from biblealignlib.burrito.util import groupby_bcv
from biblealignlib.util import DiffTargets

targetlang, sourceid = ("eng", "SBLGNT")
alset84 = AlignmentSet(targetlanguage=targetlang,
        targetid="NIV84",
        sourceid=sourceid,
        langdatapath=(CLEARROOT / f"alignments-{targetlang}/data"))
mgr84 = Manager(alset84)
niv11targets = target.TargetReader(CLEARROOT / f"alignments-{targetlang}/data/targets/NIV11/nt_NIV11.tsv")
dt84 = DiffTargets.DiffTargets84(mgr84, niv11targets, DiffTargets.BCVEQUIVALENT84)
print(f"After first pass: {len(dt84)}")
# second pass for more equivalents
dt84 = DiffTargets.DiffTargets84(mgr84, niv11targets, dt84.get_single_token_replacements())
print(f"After second pass: {len(dt84)}")
ser = DiffTargets.Serialize(dt84)
```
