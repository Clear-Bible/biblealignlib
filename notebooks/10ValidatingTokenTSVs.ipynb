# Validating Token TSV Files

In an ideal world, the source and target token corpora would be fixed once created. In practice, though, they've often changed in various ways:
- different columns/attributes (and names)
- different handling of punctuation can perturb token identifiers
- handling of Psalm titles
- extraneous content (in-line cross-references)

We need to validate token corpora files to ensure:
- tokens are correctly numbered in sequence (per verse) without numbering gaps
- tokens are well-formed: they have text, target tokens specify their `source_verse`, etc. 
- token BCVs match their versification scheme (which must therefore be specified somewhere)

(There are other kinds of validation needed in the context of alignment data: this notebook simply focuses on the token corpora.)
